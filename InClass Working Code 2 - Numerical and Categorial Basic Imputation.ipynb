### We Combine both Datasets to perform Missing Value Imputation

```python
#CodeSection6
all_data = [train,test]
```

```python
#CodeSection7
# Combine both using concat (with axis = 0) to check Missing Value
pd.concat([train,test], axis = 0).isnull().sum()
```

## 3/ Impute Numerical Variables
- Impute all Numerical Variables 

### Variable/Feature 1

```python
#CodeSection8
# Impute Feature with mean/median for both train and test - use for loop
for dataset in all_data:
  dataset['ApplicantIncome'] = dataset['ApplicantIncome'].fillna(train['ApplicantIncome'].median())
```

### Variable/Feature 2

```python
#CodeSection8
# Impute Feature with mean/median for both train and test - use for loop
for dataset in all_data:
  dataset['LoanAmount'] = dataset['LoanAmount'].fillna(train['LoanAmount'].median())
```

### Variable/Feature 3

```python
#CodeSection8
# Impute Feature with mean/median for both train and test - use for loop
for dataset in all_data:
  dataset['Loan_Amount_Term'] = dataset['Loan_Amount_Term'].fillna(train['Loan_Amount_Term'].median())
```

### Variable/Feature 4

```python
#CodeSection8
# Impute Feature with mean/median for both train and test - use for loop
for dataset in all_data:
  dataset['Credit_History'] = dataset['Credit_History'].fillna(train['Credit_History'].median())
```

## 4/ Impute Categorical Variables
- Impute all Numerical Variables

### Categorical Variable/Feature 1

```python
#CodeSection9
# Impute Feature with mode (use [0]) for both train and test - use for loop
for dataset in all_data:
  dataset['Gender'] = dataset['Gender'].fillna(train['Gender'].mode()[0])
```

### Categorical Variable/Feature 2

```python
#CodeSection9
# Impute Feature with mode (use [0]) for both train and test - use for loop
for dataset in all_data:
  dataset['Married'] = dataset['Married'].fillna(train['Married'].mode()[0])
```

### Categorical Variable/Feature 3

```python
#CodeSection9
# Impute Feature with mode (use [0]) for both train and test - use for loop
for dataset in all_data:
  dataset['Dependents'] = dataset['Dependents'].fillna(train['Dependents'].mode()[0])
```

### Categorical Variable/Feature 4

```python
#CodeSection9
# Impute Feature with mode (use [0]) for both train and test - use for loop
for dataset in all_data:
  dataset['Self_Employed'] = dataset['Self_Employed'].fillna(train['Self_Employed'].mode()[0])
```

## 5/ Now Re Check all of them

```python
#CodeSection10
# Combine both to check Missing Value
pd.concat([train, test], axis = 0).isnull().sum()
```

## 6/ Prepare Data for Model Model

```python
#CodeSection11
# Check first 5 rows of Test Dataset
test.head()
```

```python
#CodeSection12
# Divide Data into X and y
X = train.drop(['Loan_ID','Loan_Status'], axis = 1)
y = train['Loan_Status']
```

### Create Dummy Variables for  Categorical Variables

```python
#CodeSection13
# Create Dummy Variables
X = pd.get_dummies(X)
```

```python
#CodeSection14
# Check first 5 rows of X
X.head()
```

## 7/ Build Model

```python
#CodeSection15
#Start by importing LogisticRegression 
from sklearn.linear_model import LogisticRegression

#Initialize LogisticRegression
clf = LogisticRegression(max_iter=1000)
```

```python
#CodeSection16
# Fit Model
model = clf.fit(X,y)
```

### Check Accuracy of Model on Train Data

```python
#CodeSection17
# Predict on Train Data and check predictions
y_pred = model.predict(X)
y_pred
```

### We can use a Accuracy Function from Metrics

```python
#CodeSection17
# Import accuracy_score 
from sklearn.metrics import accuracy_score
```

```python
#CodeSection18
# Get Actual Predictions
y_true = train['Loan_Status']
```

```python
#CodeSection19
# Compare True values versus prediction we made
print(f" Train Accuracy : {accuracy_score(y_true, y_pred):0.1%}")
```

## 8/ Prediction and Submission
